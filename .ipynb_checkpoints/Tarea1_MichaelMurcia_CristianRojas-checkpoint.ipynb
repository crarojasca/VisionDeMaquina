Se ilustra un estudio, implementación y análisis de desempeño de las siguientes técnicas de interpolación:
<ul>
    <li> Interpolación por Vecinos más Cercanos
    <li> Interpolación Bilineal
    <li> Interpolación Bicúbica
    <li> Interpolación B-Splines
</ul>

De forma general, la interpolación en las fotografías consiste en un proceso en el cual se calculan valores numéricos desconocidos a partir de valores conocidos, esto sirve para obtener una nueva imagen a partir de una imagen base a la cual se le aplican transformaciones geométricas como escalado (Modificación del tamaño de la imagen), rotación (Aplicación de giro de una imagen en cierto ángulo), entre otras operaciones.

<h1 style="font-size:240%; color:red; text-align:center; line-height : 55px;"> 1.1. Interpolación por Vecinos más Cercanos </h1>

Es un método en el que simplemente se aumenta el tamaño de cada pixel de la imagen en cierto factor de escala. Visualmente, a partir de una imagen, esta aumentará su tamaño y se notará un efecto de pixelado.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

#Este es un paso intermedio de la interpolacion por vecinos mas cercanos, donde se aumenta
#el tamano de la imagen en un factor de escala (x2, x3, x4, ...) y los pixeles vecinos en
#la imagen original tendran color negro en la imagen resultante de este paso
def preinterpolacionVecinosMasCercanos(imagen, factorEscala):
    #Reconocimiento de las dimensiones de la imagen original (Pixeles de ancho y de alto)
    altoImagen, anchoImagen = imagen.shape[:2]
    #Creacion de una imagen en blanco ampliada
    imagenResultante = np.zeros((altoImagen * factorEscala, anchoImagen * factorEscala, 3), np.uint8)
    #Se envia cada pixel (i, j) de la imagen original al pixel (i * factorEscala, j * factorEscala)
    #de la imagen resultante, los demas pixeles se dejan de color negro para este paso.
    #Posteriormente, esos pixeles tendran el color del pixel mas cercano al pixel que se esta
    #interpolando
    for fila in range(0, altoImagen):
        for columna in range(0, anchoImagen):
            imagenResultante[fila * factorEscala][columna * factorEscala][0] = imagen[fila][columna][0]
            imagenResultante[fila * factorEscala][columna * factorEscala][1] = imagen[fila][columna][1]
            imagenResultante[fila * factorEscala][columna * factorEscala][2] = imagen[fila][columna][2]
    return imagenResultante

#Este es el ultimo paso de la interpolacion por vecinos mas cercanos, donde a partir de la
#imagen obtenida en el paso intermedio, los pixeles que no estan en una posicion de la forma
#(i * factorEscala, j * factorEscala) tendran el color del pixel mas cercano al pixel que se esta
#interpolando (Pixel (i, j))
def interpolacionVecinosMasCercanos(preimagenResultante, factorEscala):
    #Reconocimiento de las dimensiones de la imagen resultante del paso intermedio (Pixeles de
    #ancho y de alto)
    altoImagen, anchoImagen = preimagenResultante.shape[:2]
    #Creacion de una nueva imagen para no alterar la imagen obtenida del paso intermedio
    imagenResultante = preimagenResultante.copy()
    for fila in range(0, altoImagen):
        for columna in range(0, anchoImagen):
            if fila % factorEscala != 0 or columna % factorEscala != 0:
                imagenResultante[fila][columna][0] = imagenResultante[fila - (fila % factorEscala)][columna - (columna % factorEscala)][0]
                imagenResultante[fila][columna][1] = imagenResultante[fila - (fila % factorEscala)][columna - (columna % factorEscala)][1]
                imagenResultante[fila][columna][2] = imagenResultante[fila - (fila % factorEscala)][columna - (columna % factorEscala)][2]
    return imagenResultante

escala = 3
imagenOriginal = cv2.imread('Figures/imagenBase.jpg', 1)
imagenPreinterpolada = preinterpolacionVecinosMasCercanos(imagenOriginal, escala)
imagenInterpolada = interpolacionVecinosMasCercanos(imagenPreinterpolada, escala)

plt.figure(1)
plt.subplot(111)
plt.imshow(imagenOriginal)
plt.title('Imagen Original')
plt.show()

plt.figure(2)
plt.subplot(111)
plt.imshow(imagenPreinterpolada)
plt.title('Imagen Preinterpolada')
plt.show()

plt.figure(3)
plt.subplot(111)
plt.imshow(imagenInterpolada)
plt.title('Imagen Interpolada')
plt.show()
```
